## PARTE 2 ‚Äì M√©todos e Heran√ßa

**Tarefas:**
- Criar pelo menos 2 m√©todos de a√ß√£o para a classe principal.
- Criar uma subclasse herdando da classe principal.
- Adicionar pelo menos 1 novo atributo e 1 novo m√©todo exclusivo na subclasse.
- Utilizar `super()` no construtor da subclasse.

**C√≥digo:**

```python
# Desenvolva aqui o c√≥digo da Parte 2
class Servico_maria:
    def __init__(self, nome, descricao, preco):
        self.nome = nome
        self.descricao = descricao
        self.preco = preco
    def exibir_dados(self):
        print(f'Nome: {self.nome}')
        print(f'Descri√ßao: {self.descricao}')
        print(f'Pre√ßo: R${self.preco:.2f}')

    def aplicar_desconto(self, percentual):
        desconto = (self.preco * percentual) / 100
        self.preco -= desconto
        print(f"Desconto de {percentual}% aplicado. Novo pre√ßo: R${self.preco: .2f}")
    
    def atualizar_preco(self, novo_preco):
        self.preco = novo_preco
        print(f"Pre√ßo atualizado para: R${self.preco: .2f}")

class ServicoPremium_maria(Servico_maria):
  def __init__(self, nome, descricao, preco, taxa_extra):
    super().__init__(nome, descricao, preco)
    self.taxa_extra = taxa_extra

  def aplicar_taxa_extra(self):
    self.preco += self.taxa_extra
    print(f'Taxa extra aplicada. Novo pre√ßo: R${self.preco: .2f}')

  def exibir_dados(self):
    super().exibir_dados()
    print(f'Taxa extra (premium): R${self.taxa_extra: .2f}')
  

servico1 = Servico_maria('Limpeza residencial', 'Limpeza completa de casas e apartamentos', 150.00)
servico2 = Servico_maria('Manuten√ß√£o de PC', 'Formata√ß√£o, limpeza e instala√ß√£o de softwares', 200.00) 

print("Antes de aplicar a√ß√µes: ")
servico1.exibir_dados()
servico2.exibir_dados()

servico1.aplicar_desconto(10)
servico2.atualizar_preco(180.00)

print("\nAp√≥s as a√ß√µes:")
servico1.exibir_dados()
servico2.exibir_dados()

print('\nServi√ßo Premium:')
premium = ServicoPremium_maria('Limpeza p√≥s-obra', 'Limpeza pesada ap√≥s reforma', 300.00, 50.00)
premium.exibir_dados()
premium.aplicar_taxa_extra()
premium.exibir_dados()

```

## PARTE 5 ‚Äì Projeto Final e Organiza√ß√£o

**Tarefas:**
- Consolidar todas as partes do projeto em um √∫nico script organizado.
- Comentar o c√≥digo explicando cada parte.
- Testar todas as funcionalidades.

**C√≥digo:**

```python
import mysql.connector
import uuid

class DBManager:
    def __init__(self):
        self.conn = mysql.connector.connect(
            host='localhost',
            user='root',
            password='',  # ajuste sua senha
            database='servicos_db'
        )
        self.cursor = self.conn.cursor()

    def inserir_servico(self, servico):
        sql = "INSERT INTO servicos (id, nome, descricao, preco, taxa_extra) VALUES (%s, %s, %s, %s, %s)"
        valores = (servico.get_codigo_interno(), servico.nome, servico.descricao, servico.preco,
                   getattr(servico, 'taxa_extra', None))
        self.cursor.execute(sql, valores)
        self.conn.commit()

    def listar_servicos(self):
        self.cursor.execute("SELECT id, nome, descricao, preco, taxa_extra FROM servicos")
        return self.cursor.fetchall()

    def buscar_servico_por_id(self, id_servico):
        self.cursor.execute("SELECT id, nome, descricao, preco, taxa_extra FROM servicos WHERE id = %s", (id_servico,))
        return self.cursor.fetchone()

    def atualizar_servico(self, id_servico, nome, descricao, preco, taxa_extra):
        sql = """UPDATE servicos
                 SET nome=%s, descricao=%s, preco=%s, taxa_extra=%s
                 WHERE id=%s"""
        self.cursor.execute(sql, (nome, descricao, preco, taxa_extra, id_servico))
        self.conn.commit()

    def deletar_servico(self, id_servico):
        self.cursor.execute("DELETE FROM servicos WHERE id=%s", (id_servico,))
        self.conn.commit()

    def fechar_conexao(self):
        self.cursor.close()
        self.conn.close()

class Servico:
    def __init__(self, nome, descricao, preco, taxa_extra=None, id=None):
        self.nome = nome
        self.descricao = descricao
        self.preco = preco
        self.taxa_extra = taxa_extra
        self.__codigo_interno = id if id else str(uuid.uuid4())

    def get_codigo_interno(self):
        return self.__codigo_interno

# fun√ß√µes para opera√ß√µes (igual antes)...

def adicionar_servico(db):
    nome = input("Nome do servi√ßo: ").strip()
    descricao = input("Descri√ß√£o do servi√ßo: ").strip()
    preco = float(input("Pre√ßo do servi√ßo: ").strip())
    tipo = input("√â servi√ßo premium? (s/n): ").strip().lower()
    taxa_extra = None
    if tipo == 's':
        taxa_extra = float(input("Informe a taxa extra: ").strip())
    servico = Servico(nome, descricao, preco, taxa_extra)
    db.inserir_servico(servico)
    print("Servi√ßo adicionado com sucesso!")

def listar_servicos(db):
    servicos = db.listar_servicos()
    if not servicos:
        print("Nenhum servi√ßo cadastrado.")
        return
    print("\nLista de Servi√ßos:")
    for i, (id_, nome, descricao, preco, taxa_extra) in enumerate(servicos, 1):
        print(f"{i}. ID: {id_}")
        print(f"   Nome: {nome}")
        print(f"   Descri√ß√£o: {descricao}")
        print(f"   Pre√ßo: R${preco:.2f}")
        if taxa_extra is not None:
            print(f"   Taxa extra: R${taxa_extra:.2f}")
        print("")

def editar_servico(db):
    listar_servicos(db)
    id_servico = input("Digite o ID do servi√ßo que quer editar: ").strip()
    servico = db.buscar_servico_por_id(id_servico)
    if not servico:
        print("Servi√ßo n√£o encontrado.")
        return
    print("Deixe em branco para n√£o alterar o campo.")
    nome = input(f"Nome ({servico[1]}): ").strip() or servico[1]
    descricao = input(f"Descri√ß√£o ({servico[2]}): ").strip() or servico[2]
    preco_input = input(f"Pre√ßo ({servico[3]}): ").strip()
    preco = float(preco_input) if preco_input else servico[3]
    taxa_extra_input = input(f"Taxa extra ({servico[4]}): ").strip()
    taxa_extra = float(taxa_extra_input) if taxa_extra_input else servico[4]

    db.atualizar_servico(id_servico, nome, descricao, preco, taxa_extra)
    print("Servi√ßo atualizado com sucesso!")

def excluir_servico(db):
    listar_servicos(db)
    id_servico = input("Digite o ID do servi√ßo que quer excluir: ").strip()
    servico = db.buscar_servico_por_id(id_servico)
    if not servico:
        print("Servi√ßo n√£o encontrado.")
        return
    confirma = input(f"Tem certeza que deseja excluir o servi√ßo '{servico[1]}'? (s/n): ").strip().lower()
    if confirma == 's':
        db.deletar_servico(id_servico)
        print("Servi√ßo exclu√≠do com sucesso!")
    else:
        print("Opera√ß√£o cancelada.")

def menu():
    print("""
--- Sistema de Servi√ßos ---
1. Adicionar servi√ßo
2. Listar servi√ßos
3. Editar servi√ßo
4. Excluir servi√ßo
5. Sair
""")

# Aqui, n√£o vamos rodar o loop infinito autom√°tico,
# mas sim deixar o controle pro usu√°rio rodar c√©lulas ou chamar fun√ß√µes manualmente.

# Exemplo de uso:

db = DBManager()

# Para mostrar o menu e escolher uma op√ß√£o, voc√™ pode executar essa c√©lula v√°rias vezes:
menu()
opcao = input("Escolha uma op√ß√£o: ").strip()

if opcao == '1':
    adicionar_servico(db)
elif opcao == '2':
    listar_servicos(db)
elif opcao == '3':
    editar_servico(db)
elif opcao == '4':
    excluir_servico(db)
elif opcao == '5':
    print("Fechando conex√£o e saindo...")
    db.fechar_conexao()
else:
    print("Op√ß√£o inv√°lida.")
```
