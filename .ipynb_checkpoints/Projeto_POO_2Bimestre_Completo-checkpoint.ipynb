## PARTE 2 ‚Äì M√©todos e Heran√ßa

**Tarefas:**
- Criar pelo menos 2 m√©todos de a√ß√£o para a classe principal.
- Criar uma subclasse herdando da classe principal.
- Adicionar pelo menos 1 novo atributo e 1 novo m√©todo exclusivo na subclasse.
- Utilizar `super()` no construtor da subclasse.

**C√≥digo:**

```python
# Desenvolva aqui o c√≥digo da Parte 2
class Servico_maria:
    def __init__(self, nome, descricao, preco):
        self.nome = nome
        self.descricao = descricao
        self.preco = preco
    def exibir_dados(self):
        print(f'Nome: {self.nome}')
        print(f'Descri√ßao: {self.descricao}')
        print(f'Pre√ßo: R${self.preco:.2f}')

    def aplicar_desconto(self, percentual):
        desconto = (self.preco * percentual) / 100
        self.preco -= desconto
        print(f"Desconto de {percentual}% aplicado. Novo pre√ßo: R${self.preco: .2f}")
    
    def atualizar_preco(self, novo_preco):
        self.preco = novo_preco
        print(f"Pre√ßo atualizado para: R${self.preco: .2f}")

class ServicoPremium_maria(Servico_maria):
  def __init__(self, nome, descricao, preco, taxa_extra):
    super().__init__(nome, descricao, preco)
    self.taxa_extra = taxa_extra

  def aplicar_taxa_extra(self):
    self.preco += self.taxa_extra
    print(f'Taxa extra aplicada. Novo pre√ßo: R${self.preco: .2f}')

  def exibir_dados(self):
    super().exibir_dados()
    print(f'Taxa extra (premium): R${self.taxa_extra: .2f}')
  

servico1 = Servico_maria('Limpeza residencial', 'Limpeza completa de casas e apartamentos', 150.00)
servico2 = Servico_maria('Manuten√ß√£o de PC', 'Formata√ß√£o, limpeza e instala√ß√£o de softwares', 200.00) 

print("Antes de aplicar a√ß√µes: ")
servico1.exibir_dados()
servico2.exibir_dados()

servico1.aplicar_desconto(10)
servico2.atualizar_preco(180.00)

print("\nAp√≥s as a√ß√µes:")
servico1.exibir_dados()
servico2.exibir_dados()

print('\nServi√ßo Premium:')
premium = ServicoPremium_maria('Limpeza p√≥s-obra', 'Limpeza pesada ap√≥s reforma', 300.00, 50.00)
premium.exibir_dados()
premium.aplicar_taxa_extra()
premium.exibir_dados()

```

## PARTE 5 ‚Äì Projeto Final e Organiza√ß√£o

**Tarefas:**
- Consolidar todas as partes do projeto em um √∫nico script organizado.
- Comentar o c√≥digo explicando cada parte.
- Testar todas as funcionalidades.

**C√≥digo:**

```python
# Desenvolva aqui o c√≥digo da Parte 5
import mysql.connector  # Importa biblioteca para conectar ao MySQL
import uuid  # Importa biblioteca para gerar IDs √∫nicos universais

# Classe que gerencia conex√£o e opera√ß√µes no banco de dados
class DBManager:
    def __init__(self):
        # Estabelece conex√£o com banco MySQL usando host, usu√°rio e banco espec√≠ficos
        self.conn = mysql.connector.connect(
            host='localhost',
            user='root',
            database='servicos_db'
        )
        self.cursor = self.conn.cursor()  # Cria cursor para executar comandos SQL

    # M√©todo para inserir um servi√ßo na tabela 'servicos'
    def inserir_servico(self, servico):
        sql = "INSERT INTO servicos (id, nome, descricao, preco, taxa_extra) VALUES (%s, %s, %s, %s, %s)"
        # Pega os valores do servi√ßo, incluindo taxa_extra se existir, sen√£o None
        valores = (servico.get_codigo_interno(), servico.nome, servico.descricao, servico.preco,
                   getattr(servico, 'taxa_extra', None))
        self.cursor.execute(sql, valores)  # Executa o comando SQL com os valores fornecidos
        self.conn.commit()  # Confirma a opera√ß√£o no banco
        print(f"Servi√ßo '{servico.nome}' inserido no banco de dados.")  # Confirma√ß√£o na tela

    # M√©todo para fechar conex√£o e cursor do banco
    def fechar_conexao(self):
        self.cursor.close()  # Fecha o cursor
        self.conn.close()    # Fecha a conex√£o com o banco


# Classe que representa um servi√ßo comum
class Servico_maria:
    def __init__(self, nome, descricao, preco):
        self.nome = nome  # Nome do servi√ßo
        self.descricao = descricao  # Descri√ß√£o do servi√ßo
        self.preco = preco  # Pre√ßo do servi√ßo
        self.__codigo_interno = str(uuid.uuid4())  # C√≥digo interno √∫nico, privado

    # Exibe os dados do servi√ßo formatados
    def exibir_dados(self):
        print(f'Nome: {self.nome}')
        print(f'Descri√ßao: {self.descricao}')
        print(f'Pre√ßo: R${self.preco:.2f}')

    # Aplica desconto percentual no pre√ßo e atualiza valor
    def aplicar_desconto(self, percentual):
        desconto = (self.preco * percentual) / 100
        self.preco -= desconto
        print(f"Desconto de {percentual}% aplicado. Novo pre√ßo: R${self.preco: .2f}")

    # Atualiza o pre√ßo para um novo valor
    def atualizar_preco(self, novo_preco):
        self.preco = novo_preco
        print(f"Pre√ßo atualizado para: R${self.preco: .2f}")

    # Retorna o c√≥digo interno privado
    def get_codigo_interno(self):
        return self.__codigo_interno

    # Atualiza o c√≥digo interno (m√©todo setter)
    def set_codigo_interno(self, novo_codigo):
        self.__codigo_interno = novo_codigo


# Classe que representa um servi√ßo premium, herda de Servico_maria
class ServicoPremium_maria(Servico_maria):
    def __init__(self, nome, descricao, preco, taxa_extra):
        super().__init__(nome, descricao, preco)  # Chama o construtor da classe base
        self.taxa_extra = taxa_extra  # Define taxa extra para servi√ßo premium

    # Aplica taxa extra ao pre√ßo do servi√ßo
    def aplicar_taxa_extra(self):
        self.preco += self.taxa_extra
        print(f'Taxa extra aplicada. Novo pre√ßo: R${self.preco: .2f}')

    # Exibe dados do servi√ßo premium, incluindo taxa extra
    def exibir_dados(self):
        super().exibir_dados()  # Exibe dados do servi√ßo base
        print(f'Taxa extra (premium): R${self.taxa_extra: .2f}')


# Criando inst√¢ncia do gerenciador do banco
db = DBManager()

# Criando servi√ßos
servico1 = Servico_maria('Limpeza residencial', 'Limpeza completa de casas e apartamentos', 150.00)
servico2 = Servico_maria('Manuten√ß√£o de PC', 'Formata√ß√£o, limpeza e instala√ß√£o de softwares', 200.00)
servico3 = Servico_maria('Passeio com cachorro', 'Passear, alimentar e dar agua', 30.00)
servico4 = Servico_maria('Reparos automotivos', 'Reparos gerais de autom√≥veis', 250.00)
premium = ServicoPremium_maria('Limpeza p√≥s-obra', 'Limpeza pesada ap√≥s reforma', 300.00, 50.00)

# Inserindo servi√ßos no banco de dados
db.inserir_servico(servico1)
db.inserir_servico(servico2)
db.inserir_servico(servico3)
db.inserir_servico(premium)

# Fechando conex√£o com banco
db.fechar_conexao()
```
